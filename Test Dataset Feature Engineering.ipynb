```python
data=dataset.copy()
```

## Fill in Missing Categorical Features 

```python
missing_values=[feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtype=='O']
missing_values
```

```python
for feature in missing_values:
    data[feature]=data[feature].fillna('Missing')
```

```python
data.head()
```

## Converting Year Feature to Numerical Feature

```python
temporal_feature=[feature for feature in data.columns if 'Year' in feature or 'Yr' in feature]
temporal_feature
```

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    data[feature]=data['YrSold']-data[feature]
```

```python
data['GarageYrBlt'].median()
```

## Fill in Missing Numerical Features

```python
numerical_feature=[feature for feature in data.columns if data[feature].dtype!='O' and data[feature].isnull().sum()>1]
numerical_feature
```

```python
for feature in numerical_feature:
    median_value=data[feature].median()
    data[feature+'_nan']=np.where(data[feature].isnull(),1,0)
    data[feature]=np.where(data[feature].isnull(),median_value,data[feature])
data[numerical_feature].isnull().sum()    
```

## Log Transformation

```python
numerical_feature=[feature for feature in data if data[feature].dtype!='O' and feature not in temporal_feature+['Id']]
numerical_feature
```

```python
for feature in numerical_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
data.head()
```

## Handling Rare Variables

```python
categorical_feature=[feature for feature in data.columns if data[feature].dtype=='O']

for feature in categorical_feature:
    temp=data.groupby(feature)['Id'].count()/len(data)
    temp_var=temp[temp>0.01].index
    data[feature]=np.where(data[feature].isin(temp_var),data[feature],'Rare_Var')
```

```python
data.head(20)
```

## Standardization - MinMaxScaler

```python
numerical_feature=[feature for feature in data.columns if data[feature].dtype!='O' and feature not in ['Id']]
numerical_feature
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(data[numerical_feature])
```

```python
scaler.transform(data[numerical_feature])
```

```python
modified_data=pd.concat([data['Id'].reset_index(drop=True),data[categorical_feature].reset_index(drop=True),pd.DataFrame(scaler.transform(data[numerical_feature]),columns=numerical_feature)],axis=1)
modified_data.head()
```

```python
modified_data.to_csv('X_test.csv',index=False)
```
